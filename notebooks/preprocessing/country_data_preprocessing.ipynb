```python
df = pd.read_csv('C:/Users/mohdz/OneDrive/Desktop/Mini_Project/data/raw/co2_country_data.csv')
df.head()
```

```python
df = df.loc[:, ['country', 'year', 'population', 'gdp', 'co2']]
df.head()
```

```python
df = df.loc[df['year'] >= 1950, ['country', 'year', 'population', 'gdp', 'co2']]
df.set_index(df.year)
df.head()
```

```python
df.count()
```

```python
import pandas as pd

df_countries = pd.read_csv('C:/Users/mohdz/OneDrive/Desktop/Mini_Project/data/raw/country.csv')

# Create a list of country names from the second CSV file
country_list = df_countries['country'].tolist()

# Loop through the first CSV file and remove non-country rows
for index, row in df.iterrows():
    if row['country'] not in country_list:
        df.drop(index, inplace=True)

# Save the cleaned DataFrame to a new CSV file
df.to_csv('C:/Users/mohdz/OneDrive/Desktop/Mini_Project/data/processed/co2_country.csv', index=False)
print('Removed the entries which are not countries')
```
